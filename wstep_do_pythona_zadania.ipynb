ZADANIE 1: Analiza danych sprzeda≈ºowych
---------------------------------------
Dane: Miesiƒôczna sprzeda≈º produktu w PLN

sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200,
            20500, 21300, 23500, 25200]

a) Oblicz podstawowe statystyki (≈õrednia, mediana, min, max)

b) Znajd≈∫ miesiƒÖce ze sprzeda≈ºƒÖ powy≈ºej ≈õredniej

c) Oblicz procentowy wzrost sprzeda≈ºy (grudzie≈Ñ vs stycze≈Ñ)

d) Stw√≥rz wykres liniowy pokazujƒÖcy trend

e) Dodaj liniƒô ≈õredniej na wykresie


````python
# Dane sprzeda≈ºy
sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200, 20500, 21300, 23500, 25200]
miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze', 'Lip', 'Sie', 'Wrz', 'Pa≈∫', 'Lis', 'Gru']

# --- a) Podstawowe statystyki ---
srednia = np.mean(sprzedaz)
mediana = np.median(sprzedaz)
minimum = np.min(sprzedaz)
maksimum = np.max(sprzedaz)

print("=== a) Podstawowe statystyki ===")
print(f"≈örednia: {srednia:.2f} PLN")
print(f"Mediana: {mediana:.2f} PLN")
print(f"Min: {minimum} PLN")
print(f"Maks: {maksimum} PLN\n")

# --- b) MiesiƒÖce ze sprzeda≈ºƒÖ powy≈ºej ≈õredniej ---
powyzej_sredniej = [miesiace[i] for i, s in enumerate(sprzedaz) if s > srednia]
print("=== b) MiesiƒÖce powy≈ºej ≈õredniej ===")
print(powyzej_sredniej, "\n")

# --- c) Procentowy wzrost sprzeda≈ºy (grudzie≈Ñ vs stycze≈Ñ) ---
wzrost_proc = ((sprzedaz[-1] - sprzedaz[0]) / sprzedaz[0]) * 100
print("=== c) Wzrost sprzeda≈ºy (grudzie≈Ñ vs stycze≈Ñ) ===")
print(f"{wzrost_proc:.2f}%\n")

# --- d) Wykres liniowy trendu ---
plt.figure()
plt.plot(miesiace, sprzedaz, marker='o', color='blue')
plt.title("üìà Trend sprzeda≈ºy produktu")
plt.xlabel("MiesiƒÖc")
plt.ylabel("Sprzeda≈º (PLN)")
plt.grid(True)
plt.show()

# --- e) Wykres z liniƒÖ ≈õredniej ---
plt.figure()
plt.plot(miesiace, sprzedaz, marker='o', color='blue', label='Sprzeda≈º')
plt.axhline(y=srednia, color='red', linestyle='--', label=f'≈örednia = {srednia:.0f} PLN')
plt.title("üìâ Trend sprzeda≈ºy z liniƒÖ ≈õredniej")
plt.xlabel("MiesiƒÖc")
plt.ylabel("Sprzeda≈º (PLN)")
plt.legend()
plt.grid(True)
plt.show()
````

ADANIE 2: S≈Çowniki - Analiza portfolio
---------------------------------------
Stw√≥rz s≈Çownik opisujƒÖcy portfolio inwestycyjne:
```
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}
```

a) Oblicz ca≈ÇkowitƒÖ warto≈õƒá poczƒÖtkowƒÖ portfolio

b) Oblicz aktualnƒÖ warto≈õƒá portfolio

c) Oblicz zysk/stratƒô dla ka≈ºdej akcji

d) Kt√≥re akcje przynoszƒÖ najwiƒôkszy zwrot (w %)?


````python
# --- Dane: s≈Çownik opisujƒÖcy posiadane akcje ---
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}

# --- a) Oblicz warto≈õƒá zakupu, warto≈õƒá aktualnƒÖ i zysk dla ka≈ºdej akcji ---
print("=== a) Szczeg√≥≈Çy portfela ===")
for ticker, dane in portfolio.items():
    liczba = dane['liczba']
    zakup = dane['cena_zakupu']
    aktualna = dane['cena_aktualna']
    koszt = liczba * zakup
    wartosc = liczba * aktualna
    zysk = wartosc - koszt

    # Dodaj zysk do s≈Çownika
    dane['koszt'] = koszt
    dane['wartosc'] = wartosc
    dane['zysk'] = zysk

    print(f"{ticker}: zakup = {koszt} USD, aktualna = {wartosc} USD, zysk = {zysk} USD")
print()

# --- b) Ca≈Çkowity zysk portfela ---
caly_zysk = sum(d['zysk'] for d in portfolio.values())
print("=== b) Ca≈Çkowity zysk portfela ===")
print(f"≈ÅƒÖczny zysk: {caly_zysk} USD\n")

# --- c) Oblicz procentowy zwrot (ROI) dla ka≈ºdej akcji ---
print("=== c) Procentowy zwrot (ROI) ===")
for ticker, dane in portfolio.items():
    roi = (dane['zysk'] / dane['koszt']) * 100
    dane['ROI'] = roi
    print(f"{ticker}: ROI = {roi:.2f}%")
print()

# --- d) Znajd≈∫ akcjƒô z najwy≈ºszym ROI ---
najlepsza = max(portfolio.items(), key=lambda x: x[1]['ROI'])
print("=== d) Najlepsza inwestycja ===")
print(f"{najlepsza[0]} z ROI = {najlepsza[1]['ROI']:.2f}%")
````

ZADANIE 3: Wizualizacja danych klimatycznych
--------------------------------------------
temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]


temp_2023 = [3, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 4]

miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze',
            'Lip', 'Sie', 'Wrz', 'Pa≈∫', 'Lis', 'Gru']

a) Stw√≥rz wykres liniowy por√≥wnujƒÖcy temperatury w obu latach

b) Zaznacz najcieplejszy i najzimniejszy miesiƒÖc ka≈ºdego roku

c) Oblicz ≈õredniƒÖ r√≥≈ºnicƒô temperatur miƒôdzy latami

d) Dodaj odpowiednie etykiety, legendƒô i tytu≈Ç

````python
temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]
temp_2023 = [3, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 4]
miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze', 'Lip', 'Sie', 'Wrz', 'Pa≈∫', 'Lis', 'Gru']

# --- a) Wykres por√≥wnawczy ---
plt.figure()
plt.plot(miesiace, temp_2022, marker='o', label='2022', color='blue')
plt.plot(miesiace, temp_2023, marker='o', label='2023', color='orange')
plt.title("üåç Por√≥wnanie ≈õrednich temperatur 2022 vs 2023")
plt.xlabel("MiesiƒÖc")
plt.ylabel("Temperatura [¬∞C]")
plt.legend()
plt.grid(True)
plt.show()

# --- b) Najcieplejszy i najzimniejszy miesiƒÖc ---
najcieplejszy_2022 = miesiace[np.argmax(temp_2022)]
najzimniejszy_2022 = miesiace[np.argmin(temp_2022)]
najcieplejszy_2023 = miesiace[np.argmax(temp_2023)]
najzimniejszy_2023 = miesiace[np.argmin(temp_2023)]

print("=== b) Najcieplejsze i najzimniejsze miesiƒÖce ===")
print(f"2022: üî• {najcieplejszy_2022}, ‚ùÑÔ∏è {najzimniejszy_2022}")
print(f"2023: üî• {najcieplejszy_2023}, ‚ùÑÔ∏è {najzimniejszy_2023}\n")

# --- c) ≈örednia r√≥≈ºnica temperatur ---
roznice = np.array(temp_2023) - np.array(temp_2022)
srednia_roznica = np.mean(roznice)
print("=== c) ≈örednia r√≥≈ºnica temperatur miƒôdzy latami ===")
print(f"{srednia_roznica:.2f}¬∞C\n")

# --- d) Wykres z zaznaczonymi ekstremami ---
plt.figure()
plt.plot(miesiace, temp_2022, marker='o', label='2022', color='blue')
plt.plot(miesiace, temp_2023, marker='o', label='2023', color='orange')
plt.scatter(najcieplejszy_2023, max(temp_2023), color='red', s=100, label='Najcieplejszy 2023')
plt.scatter(najzimniejszy_2023, min(temp_2023), color='cyan', s=100, label='Najzimniejszy 2023')
plt.title("üå§Ô∏è ≈örednie temperatury miesiƒôczne (2022 vs 2023)")
plt.xlabel("MiesiƒÖc")
plt.ylabel("Temperatura [¬∞C]")
plt.legend()
plt.grid(True)
plt.show()
````

ZADANIE 4: Symulacja Monte Carlo
---------------------------------
Zasymuluj 1000 rzut√≥w kostkƒÖ (6-≈õciennƒÖ):

a) Wygeneruj dane u≈ºywajƒÖc np.random.randint(1, 7, 1000)

b) Stw√≥rz histogram pokazujƒÖcy czƒôsto≈õƒá ka≈ºdej liczby

c) Oblicz teoretyczne i empiryczne prawdopodobie≈Ñstwo ka≈ºdej liczby

d) Czy wyniki sƒÖ zgodne z teoriƒÖ? (ka≈ºda liczba ~16.67%)

````python
# --- a) Generowanie 1000 rzut√≥w ---
rzuty = np.random.randint(1, 7, 1000)

# --- b) Histogram czƒôsto≈õci ---
plt.figure()
plt.hist(rzuty, bins=np.arange(0.5, 7.5, 1), rwidth=0.8, color='lightblue', edgecolor='black')
plt.title("üé≤ Rozk≈Çad wynik√≥w (1000 rzut√≥w kostkƒÖ)")
plt.xlabel("Wynik na kostce")
plt.ylabel("Czƒôsto≈õƒá")
plt.grid(True)
plt.show()

# --- c) Prawdopodobie≈Ñstwa empiryczne i teoretyczne ---
unikalne, liczby = np.unique(rzuty, return_counts=True)
p_empiryczne = liczby / len(rzuty)
p_teoretyczne = 1/6

print("=== c) Prawdopodobie≈Ñstwa ===")
for i, p in zip(unikalne, p_empiryczne):
    print(f"Liczba {i}: empiryczne = {p:.3f}, teoretyczne = {p_teoretyczne:.3f}")

# --- d) Ocena zgodno≈õci ---
odchylenie = np.mean(np.abs(p_empiryczne - p_teoretyczne)) * 100
print(f"\n=== d) Ocena zgodno≈õci z teoriƒÖ ===")
print(f"≈örednie odchylenie: {odchylenie:.2f}%")
if odchylenie < 2:
    print("‚úÖ Wyniki sƒÖ zgodne z teoriƒÖ (r√≥wny rozk≈Çad).")
else:
    print("‚ö†Ô∏è Niewielkie odchylenia ‚Äî naturalny efekt losowo≈õci.")
````
