```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
import os


path = 'train.csv'
df = pd.read_csv(path)

print(f"Wczytano df shape: {df.shape}")

# === Komórka 1: podstawowe informacje ===
print('\n--- Statystyki opisowe (liczbowe) ---')
display(df.describe())

print('\n--- Przykładowe wiersze ---')
display(df.head())


# === Komórka 2: dystrybucja ceny  =====
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['SalePrice'], kde=True)
plt.title('SalePrice - rozkład (oryginalny)')


# === Komórka 3: wykresy zależności dla najważniejszych zmiennych ===
plt.figure(figsize=(8,5))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title('GrLivArea vs SalePrice')
plt.show()

# 4
plt.figure(figsize=(10,5))
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.title('OverallQual vs SalePrice')
plt.show()

# 5 Średnia cena
neigh = df.groupby('Neighborhood')['SalePrice'].median().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=neigh.head(15).index, y=neigh.head(15).values)
plt.xticks(rotation=45)
plt.title('Mediana SalePrice - top 15 Neighborhood')
plt.ylabel('Mediana SalePrice')
plt.tight_layout()
plt.show()

# 6 Sprzedane domy w poszczególnych miesiącach

df['SaleDate'] = pd.to_datetime(df['YrSold'].astype(str) + '-' + df['MoSold'].astype(str))
sales_over_time = df.groupby('SaleDate').size().sort_index()
plt.figure(figsize=(12,5))
plt.plot(sales_over_time.index, sales_over_time.values)
plt.title('Trend sprzedaży domów w czasie')
plt.xlabel('Rok i miesiąc sprzedaży')
plt.ylabel('Liczba transakcji')
plt.tight_layout()
plt.show()
```
